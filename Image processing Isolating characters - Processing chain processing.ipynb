```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Chargement de l'image
image = cv2.imread('/content/text1.png')

# Affichage de l'image originale
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Image Originale')
plt.show()

# Etape 1: Pr-traitements avec filtre mdian et seuillage
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 5)

# Appliquer un filtre bilatral
#gray = cv2.bilateralFilter(gray, 9, 75, 75)

# Appliquer un filtre gaussien
#gray = cv2.GaussianBlur(gray, (5, 5), 0)

# Appliquer un filtre moyen
#gray = cv2.blur(gray, (5, 5))

kev, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)



# Affichage de l'image aprs pr-traitements
plt.imshow(thresh , cmap='gray')
plt.title('Image aprs pr-traitements (Filtre mdian avec seuillage)')
plt.show()

# Etape 2: Identification des composantes connexes par la segmentation par seuillage
contours, _ = cv2.findContours(thresh , cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Cration d'une copie de l'image pour dessiner les contours
image_contours = image.copy()

# Dessiner les contours pour localiser les caractres
cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 3)

# Affichage de l'image avec les contours
plt.imshow(cv2.cvtColor(image_contours, cv2.COLOR_BGR2RGB))
plt.title('Image aprs localisation des composantes connexes par segmentation avec seuillage')
plt.show()

# Etape 3: Post-traitements (dilatation et comptage des composantes connexes)
kernel = np.ones((3, 3), np.uint8)
dilated_thresh = cv2.dilate(thresh , kernel, iterations=1)
_, labels_dilated, stats_dilated, _ = cv2.connectedComponentsWithStats(dilated_thresh)

# Affichage de l'image aprs dilatation
plt.imshow(dilated_thresh, cmap='gray')
plt.title('Image aprs dilatation')
plt.show()

component_count_dilated = len(stats_dilated) + 5
print("Nombre de composantes connexes identifies aprs dilatation : ", component_count_dilated)
```
